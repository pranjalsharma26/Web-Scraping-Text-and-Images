```python
#import the requried libraries
import nltk
import urllib
import re
import bs4 as bs
from nltk.corpus import stopwords
nltk.download("stopwords") 
nltk.download('punkt') # for converting text into sentences
```

```python
#Getting the data source- Adding the URL
data = urllib.request.urlopen("https://www.investopedia.com/terms/m/machine-learning.asp").read()
```

```python
data
```

```python
# Parsing the data while converting them into object
soup = bs.BeautifulSoup(data, 'lxml')
```

```python
soup
```

```python
#Fetching The Data
text = ""
#Picking the contents which are inside the p tag. Example- <p>Pick Me</p>
for paragraph in soup.find_all("p"):  
    text += paragraph.text
```

```python
text
```

```python
#Pre-Porcessing the data -> Removing "\n,+,-,0-9 for [1],blankspace, lowering the text etc"
text = re.sub(r'\[[0-9]*\]',' ', text)
text = re.sub(r'\s+',' ', text)
text = text.lower()
text = re.sub(r'\d',' ', text)
```

```python
text
```

```python
# Preparing the dataset - By converting text into sentences
sentences = nltk.sent_tokenize(text)
sentences
```

```python
# Tokenizing sentences inot words
words = [nltk.word_tokenize(sentence) for sentence in sentences]
```
